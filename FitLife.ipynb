changing col type

```python
category = ['gender', 'activity_type', 'intensity', 'participant_id', 'smoking_status', 'health_condition']
df[category] = df[category].astype('category')
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

Filling null values as 'No Health Conditions'

```python
df['health_condition'] = df['health_condition'].cat.add_categories('No Health Condition')
df['health_condition'] = df['health_condition'].fillna('No Health Condition')
```

Grouping by Gender to get insights

```python
numeric_col = df.select_dtypes(include = ['number']).columns
df_gender = df.groupby('gender', observed = True)[numeric_col].agg('mean')
df_gender['gender_count'] = df.groupby('gender', observed = True)['gender'].size()
```

```python
df_gender.T
```

Visualization of age distribution

```python
sns.histplot(x = 'age', data = df, binwidth = 7, kde = True, line_kws = {'linestyle' : '--'})
plt.title('Histogram of Age Distribution', fontweight = 'bold', fontsize = 18)
plt.xlabel('Age', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.xticks(np.arange(15, 66, 5))
plt.show()
```

```python
plt.pie(df_gender['gender_count'], autopct = '%1.1f%%', shadow = True, startangle = 270, explode = [0, 0, 0.2], labels = df_gender.index)
plt.title("Pie Chart of Gender Count", fontweight = 'bold', fontsize = 16)
plt.legend(bbox_to_anchor = (0, 1))
plt.show()
```

Grouping by participant id to get insights

```python
df_id = df.groupby(['participant_id', 'gender'], observed = True)[numeric_col].agg('mean').reset_index()
df_countid = df.groupby(['participant_id', 'gender'], observed = True).size().reset_index(name = 'count')
df_id  = pd.merge(df_id, df_countid)
df_id
```

```python
df_id.T
```

```python
df.columns
```

```python
df_5id = df_id.nlargest(5, 'calories_burned')
df_5id.index = np.arange(1, 6)
df_5id['participant_id'] = df_5id['participant_id'].astype('str')
df_5id
```

```python
sns.barplot(x = 'participant_id', y = 'calories_burned', data = df_5id, hue = 'duration_minutes', color = 'blue', palette = 'deep')
plt.ylim(24, 26.5)
plt.title("Bar Diagram of top 5 individuals with highest avg calories burned", fontweight = 'bold', fontsize = 16)
plt.xlabel("Participant's id", fontsize = 14)
plt.ylabel("Average Calories Burned", fontsize = 14)
plt.legend(title = 'Avg Duration(in mins)', bbox_to_anchor = (1,1), title_fontsize = 12)
plt.show()
```

Grouping by month to get insights

```python
df_date = df.groupby(df['date'].dt.to_period('M'))[numeric_col].agg('mean').reset_index()
df_datecount = df.groupby(df['date'].dt.to_period('M')).size().reset_index(name = 'count')
df_date = pd.merge(df_date, df_datecount, on = 'date')
df_date.T
```

```python
df_count = df_date.nlargest(5, 'count')
cols = ['date', 'count'] + [col for col in df_date if col not in ['date', 'count']]
df_count = df_count[cols]
df_count
```

```python
sns.barplot(x = 'date', y = 'count', data = df_count, label = 'Number of trials took part in that month',color = 'cyan', edgecolor = 'black')
plt.title("Bar Chart of Top 5 months with most sessions", fontweight = 'bold', fontsize = 16)
plt.xlabel("Months", fontsize = 14)
plt.ylabel("Count of trials", fontsize = 14)
plt.xticks(rotation = 90)
plt.ylim(60000, 72000)
plt.legend(loc = 'upper right', bbox_to_anchor = (1, 1))
plt.show()
```

```python
sns.barplot(y = 'date', x = 'count', data = df_date, 
            label = 'Number of trials took part in that month',color = 'green', orient = 'h')
plt.title("Bar Chart of dif months with count of sessions", fontweight = 'bold', fontsize = 16)
plt.xlabel("Months", fontsize = 14)
plt.ylabel("Count of trials", fontsize = 14, labelpad = 12)
plt.legend(bbox_to_anchor = (1, 1))
plt.show()
```

```python
date_df = df.groupby('date').agg({'calories_burned' : 'mean'}).reset_index()
date_df
```

```python
plt.figure(figsize = (12, 7))
sns.lineplot(x = 'date', y = 'calories_burned', data = date_df, 
            label = 'Avg Calories burned per day',color = 'darkblue')
plt.title("Line Chart of avg calories burned per day over a year", fontweight = 'bold', fontsize = 16)
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Avg Calories Burned", fontsize = 14, labelpad = 12)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.legend(bbox_to_anchor = (1, 1))
plt.grid()
plt.show()
```

Grouping by months and gender to draw insights

```python
df_dateandgender = df.groupby([df['date'].dt.to_period('M'), 'gender'], observed = True)[numeric_col].agg('mean').reset_index()
count_df = df.groupby([df['date'].dt.to_period('M'), 'gender'], observed=True).size().reset_index(name='count')
df_dateandgender = pd.merge(df_dateandgender, count_df, on=['date', 'gender'])
df_dateandgender.T
```

```python
df_dateandgender = df_dateandgender.astype({'date' : 'str'})

plt.figure(figsize = (12, 7))
sns.lineplot(x = 'date', y = 'daily_steps', data = df_dateandgender, marker = 'o', hue = 'gender')
plt.title("Line chart of dif months with gender wise for daily steps", fontweight = 'bold', fontsize = 16)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Daily Steps", fontsize = 14)
plt.legend(bbox_to_anchor = (1, 1))
plt.grid()
plt.show()
```

Grouping by activity to draw insights

```python
df_act = df.groupby('activity_type', observed = True).agg({'activity_type' : 'count', **{col : 'mean' for col in numeric_col}})
df_act = df_act.rename(columns = {'activity_type' : 'activity_count'})
df_act = df_act.reset_index()
df_act
```

```python
sns.countplot(x = 'activity_type', data = df, label = 'count', edgecolor = 'black', color = 'turquoise')
plt.title("Count plot of dif activity types", fontweight = 'bold', fontsize = 16)
plt.xlabel("Activity Type", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
plt.legend(bbox_to_anchor = (1, 1))
plt.xticks(rotation = 90)
plt.ylim(66000, 70000)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.boxplot(x = 'activity_type', y = 'calories_burned', data = df)
plt.xticks(rotation = 90)
plt.title("Box plot distribution of calorie burned with activity type", fontweight = 'bold', fontsize = 16)
plt.xlabel("Activity Type", fontsize = 14)
plt.ylabel("Calories Burned", fontsize = 14)
plt.legend(['calories burned'], bbox_to_anchor = (1, 1))
plt.grid()
plt.yticks(np.arange(0, 101, 10))
plt.show()
```

Drawing insights of activity frequency with time

```python
df_timeact = df.groupby('date').agg({'activity_type' : 'count'}).reset_index()
df_timeact
```

```python
plt.figure(figsize = (12, 7))

sns.lineplot(x = 'date', y = 'activity_type', data = df_timeact, label = 'no. of sessions')
plt.title("Line chart of activity frequency wrt time", fontsize = 16, fontweight = 'bold')
plt.xlabel("Date", fontsize = 14)
plt.ylabel("No. of sessions", fontsize = 14)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.legend(bbox_to_anchor = (1, 1))
plt.grid()
plt.show()
```

Bar plot of count of health conditions

```python
health = df['health_condition'].value_counts()
sns.barplot(x = health.index, y = health.values, color = 'lime', edgecolor = 'black', label = 'count')
plt.title("Bar Chart showing count of dif health condition", fontweight = 'bold', fontsize = 16)
plt.xlabel("Health Condition", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
plt.legend(bbox_to_anchor = (1, 1))
plt.grid(axis = 'y')
plt.show()
```
