Grouping by buyer gender to extract insights

```python
cols = ['buyer_age', 'sales_price', 'shipping_charges', 'sales_per_unit', 'total_sales', 'quantity', 'rating']
df.groupby(by = 'buyer_gender')[cols].describe().T
```

```python
agg_fns = {'buyer_gender' : 'count', 'buyer_age' : 'mean', 'sales_price' : ['mean', 'sum'],'shipping_charges' : ['mean', 'sum'], 'sales_per_unit' : ['mean','sum'],'total_sales' : ['mean', 'sum'], 'quantity' : ['mean','sum'], 'rating' : 'mean'}
df_sex = df.groupby(by = 'buyer_gender').agg(agg_fns)
```

```python
df_sex
```

```python
plt.pie(df_sex['buyer_gender']['count'], labels = df_sex.index, autopct = '%1.1f%%', startangle = 90, shadow = True, explode = [0, 0.08])
plt.title("Comparison chart of Buyer's gender", fontweight = 'bold')
plt.axis('equal')
plt.legend(bbox_to_anchor = (0,1))
plt.show()
```

```python
sns.histplot(x = 'buyer_age', data = df, bins = 7, kde = True, line_kws = {'color' : 'blue', 'linestyle' : ':'}, color = 'red')
plt.title("Buyer's Age distribution", fontweight = 'bold')
plt.legend(['KDE line', 'Histogram bar'], loc = 'upper left', bbox_to_anchor = (1, 1))
plt.xlabel("Buyer's Age")
plt.show()
```

```python
sns.barplot(x = df_sex.index, y = df_sex['sales_price']['sum'], edgecolor = 'black')
plt.title("Buyer's Gender vs Sales Price", fontweight = 'bold')
plt.xlabel("Buyer's Gender")
plt.ylabel("Sales Price")
plt.show()
```

```python
sns.barplot(x = df_sex.index, y = df_sex['sales_per_unit']['sum'], edgecolor = 'black')
plt.title("Buyer's Gender vs Sales Per unit(sales + shipping charges)", fontweight = 'bold')
plt.xlabel("Buyer's Gender")
plt.ylabel("Sales Per unit(sales + shipping charges)")
plt.show()
```

```python
sns.barplot(x = df_sex.index, y = df_sex['total_sales']['sum'], edgecolor = 'black')
plt.title("Buyer's Gender vs Total Sales", fontweight = 'bold')
plt.xlabel("Buyer's Gender")
plt.ylabel("Total Sales")
plt.show()
```

Group by product category to get insights

```python
df.groupby(by = 'product_category').describe(include = ['int', 'float']).T
```

```python
agg_fns = {'product_category' : 'count', 'buyer_age' : 'mean', 'sales_price' : ['mean', 'sum'],'shipping_charges' : ['mean', 'sum'], 'sales_per_unit' : ['mean','sum'],'total_sales' : ['mean', 'sum'], 'quantity' : ['mean','sum'], 'rating' : 'mean'}
df_cat = df.groupby(by = 'product_category').agg(agg_fns)
```

```python
df_cat
```

```python
plt.pie(df_cat['product_category']['count'], autopct = '%1.1f%%', shadow = True, startangle = 180, explode = [0, 0, 0.14], labels = df_cat.index)
plt.title("Distribution of dif product categories", fontweight = 'bold')
plt.legend(bbox_to_anchor = (0,1))
plt.axis('equal')
plt.show()
```

```python
sns.barplot(x = df_cat.index, y = df_cat['sales_price']['sum'], color = 'orchid', edgecolor = 'black')
plt.title("Product Categoris vs Sales Price", fontweight = 'bold')
plt.xlabel("Product Categories")
plt.ylabel("Sales Price")
plt.show()
```

```python
sns.barplot(x = df_cat.index, y = df_cat['sales_per_unit']['sum'], color = 'orchid', edgecolor = 'black')
plt.title("Product Categoris vs Sales Per Unit(Sales + Shipping charges)", fontweight = 'bold')
plt.xlabel("Product Categories")
plt.ylabel("Sales Per Unit(Sales + Shipping charges)")
plt.show()
```

```python
sns.barplot(x = df_cat.index, y = df_cat['total_sales']['sum'], color = 'orchid', edgecolor = 'black')
plt.title("Product Categoris vs Total Sales", fontweight = 'bold')
plt.xlabel("Product Categories")
plt.ylabel("Total Sales")
plt.show()
```

Grouping by Monthly date of order dates

```python
df.groupby(pd.Grouper(key = 'order_date', freq = 'ME'))[cols].describe().T
```

```python
agg_fns = {"buyer_age" : "mean", "sales_price" : 'sum', "shipping_charges" : "sum", "sales_per_unit" : "sum", 'total_sales' : 'sum', 'quantity' : 'sum', 'rating' : 'mean'}
df_date = df.groupby(pd.Grouper(key = 'order_date', freq = "ME")).agg(agg_fns)
```

```python
df_date
```

```python
plt.figure(figsize = (10, 5))

sns.lineplot(x = df_date.index, y = 'sales_price', data = df_date, color = 'darkblue', label = 'Sales Price', marker = 'o')
plt.title("Sales Price for dif months", fontweight = 'bold', fontsize = 20)
plt.xlabel("Order Date", labelpad = 15, fontsize = 18)
plt.ylabel("Sales Price", labelpad = 13, fontsize = 18)
plt.xticks(df_date.index, rotation = 90)
plt.grid()
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```

```python
plt.figure(figsize = (10, 5))

sns.lineplot(x = df_date.index, y = 'sales_per_unit', data = df_date, color = 'darkblue', label = 'Sales Per Unit', marker = 'o')
plt.title("Sales Per Unit(Sales + Shipping Charges) for dif months", fontweight = 'bold', fontsize = 18)
plt.xlabel("Order Date", labelpad = 15, fontsize = 16)
plt.ylabel("Sales Per Unit(Sales + Shipping Charges)", labelpad = 13, fontsize = 16)
plt.xticks(df_date.index, rotation = 90)
plt.grid()
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```

```python
plt.figure(figsize = (10, 5))

sns.lineplot(x = df_date.index, y = 'total_sales', data = df_date, color = 'darkblue', label = 'Total Sales', marker = 'o')
plt.title("Total Sales for dif months", fontweight = 'bold', fontsize = 18)
plt.xlabel("Order Date", labelpad = 15, fontsize = 16)
plt.ylabel("Total Sales", labelpad = 13, fontsize = 16)
plt.xticks(df_date.index, rotation = 90)
plt.grid()
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```
