Q.49 Calculate the percentage increase in Salary for each employee compared to the average Salary in 
their Department. 

```python
avg_salary = df.groupby("Department", as_index = False)['Salary'].mean()
```

```python
avg_salary = avg_salary.rename(columns = {"Salary" : "Average_salary"})
avg_salary
```

```python
df_avg_salary = df.merge(avg_salary, on = "Department")
df_avg_salary.head()
```

```python
df_avg_salary['Percent_increase'] = (df_avg_salary['Salary'] - df_avg_salary['Average_salary']) / (df_avg_salary['Average_salary']) * 100
```
